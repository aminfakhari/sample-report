This report shows how to combine text, mathematics, code, and figures in a Jupyter Notebook. As an example, we will analyze a quadratic function:


$$ f(x) = ax^2 + bx + c $$

```python
# 2. Python Code – Function Definition
import numpy as np
import matplotlib.pyplot as plt

# Quadratic coefficients
a, b, c = 1, -3, 2

# Define function
def f(x):
    return a*x**2 + b*x + c

# Compute values
x = np.linspace(-1, 4, 100)
y = f(x)
```

Below is a visualization.

```python
# 3. Visualization
plt.figure(figsize=(6,4))
plt.plot(x, y, label="f(x) = x² - 3x + 2")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)

plt.title("Quadratic Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
